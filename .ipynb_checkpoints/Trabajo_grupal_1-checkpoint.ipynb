```python
# estrategias de estandarizacin que nos permitirn que las variables tengan una escala comparable,
# lo cual facilita la comparacin de sus distribuciones en el grfico
```

```python
from sklearn.preprocessing import StandardScaler

#herramienta que transformar los datos para que tengan una media de 0 y una desviacin estndar de 1.
scaler = StandardScaler()

#calcula la media y desviacin estndar en DataNames y transforma los datos
normalized_data = scaler.fit_transform(theMapAndData[DataNames])
# fit_transform calcula la media y desviacin estndar de las variables y los utiliza para transformar los datos

#nueva dataframe con datos normalizados, se pasa de ancho a largo y lo grafico en escala no logaritmica
sns.displot(pd.melt(pd.DataFrame(normalized_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
# tcnica de escalado Min-Max, a diferencia del anterior, este tambin preserva la forma original de
# la distribucin de los datos, pero en un rango especfico de 0 a 1, en este caso de -0.2 a 1.2
from sklearn.preprocessing import MinMaxScaler

# valor mnimo se convierta en 0 y el valor mximo se convierta en 1.
scaler = MinMaxScaler(feature_range=(0, 1))

#calcular el rango mnimo y mximo de las columnas en DataNames y transformar
scaled_data=scaler.fit_transform(theMapAndData[DataNames])

#grafico de distribucion de un data frame nuevo con los datos transformados del DataNames pasado de ancho a largo
sns.displot(pd.melt(pd.DataFrame(scaled_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
# tcnica de escalado robusto
from sklearn.preprocessing import RobustScaler
#En si todo el proceso normaliza las variables, haciendo menos susceptibles a influencias de valores atpicos.

#transformar los datos de manera que los valores estn centrados en la mediana y se escalen segn el rango intercuartlico
scaler = RobustScaler()

#calcular la mediana y el rango intercuartlico de las columnas en DataNames y transforma datos
robScaled_data = scaler.fit_transform(theMapAndData[DataNames])

#grafico de distribucion de un data frame nuevo con los datos transformados del DataNames pasado de ancho a largo
sns.displot(pd.melt(pd.DataFrame(robScaled_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
#transformacin de cuantiles:
from sklearn.preprocessing import QuantileTransformer
#En si transforma cuantiles a las variables, ajustndolas a una distribucin normal.

#Los datos sern transformados en 99 rangos diferentes, se transformarn a una distribucin normal/uniforme
scaler = QuantileTransformer(n_quantiles=99, random_state=0,output_distribution='normal')

#datos transformados en la distribucin normal especificada.
QtScaled_data = scaler.fit_transform(theMapAndData[DataNames])

#Grafica distribucion de un data frame nuevo con los datos transformados del DataNames pasado de ancho a largo
sns.displot(pd.melt(pd.DataFrame(QtScaled_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
#Nos quedamos con QtScaled_data (el ultimo)
theMapAndData['fragility_Qt']=QtScaled_data[:,0]
```

```python

import mapclassify #mapclassify para generar diferentes esquemas de clasificacin segn intervalos
import numpy as np

np.random.seed(12345) # so we all get the same results!
#para 5 intervalos, theVar asigno la columna fragility_Qt de theMapAndData
K=5
theVar=theMapAndData.fragility_Qt #esta es la fragilityqt
#EqualInterval---->Divide los valores de la variable en intervalos de igual tamao
ei5 = mapclassify.EqualInterval(theVar, k=K)
#StdMean---->Crea intervalos basados en la media y la desviacin estndar
msd = mapclassify.StdMean(theVar)
#Quantiles--->Crea intervalos donde cada uno contiene aproximadamente la misma cantidad de observaciones
q5=mapclassify.Quantiles(theVar,k=K)

#Basado en similitud, bueno para data multimodales
#detecta los puntos de ruptura entre los grupos de datos
mb5 = mapclassify.MaximumBreaks(theVar, k=K)
# Basado en similitud, bueno para datos asimetricos/asesgados
ht = mapclassify.HeadTailBreaks(theVar) # no K needed
# Basado en similitud, optimiza la homogeneidad dentro de los intervalos
# maximiza las diferencias entre rangos
fj5 = mapclassify.FisherJenks(theVar, k=K)
# Basado en similitud, maximiza las diferencias entre rangos
jc5 = mapclassify.JenksCaspall(theVar, k=K)
#Basado en similitud, Maximiza la homogeneidad dentro de los intervalos para datos sesgado
mp5 = mapclassify.MaxP(theVar, k=K)
```

```python
#lista de los diferentes metodos de clasificacin de intervalos de la variable TheVar
class5 = ei5,msd, q5,mb5,  ht, fj5, jc5, mp5
#matriz de valores de ADCM de cada clasificador en class5.
#ADCM mide la desviacin absoluta desde la mediana de cada intervalo
fits = np.array([ c.adcm for c in class5])
# pasamos ADCM a dataFrame
adcms = pd.DataFrame(fits)
# add columna classifier de adcms
adcms['classifier'] = [c.name for c in class5]
# Add column names to the ADCM
adcms.columns = ['ADCM', 'Classifier']
```

```python
#muestro el grafico generado
adcms.sort_values('ADCM').plot.barh(x='Classifier')
# ADCM significa calculo de la desviacion estandar, el grafico es sd de cada tipo de agrupacion de intervalo q se ha dado
```

```python
# guardamos el mejor mtodo de clasificacin en una nueva columna llamada fragility_Qt_jc5
theMapAndData['fragility_Qt_jc5'] = jc5.yb
```

```python
# datos originales con fragility_Qt y datos clasificados con fragility_Qt_jc5
theMapAndData[['fragility_Qt','fragility_Qt_jc5']].head()
```

```python
# este cdigo tiene como objetivo general la media de fragility_Qt para cada categoria fragility_Qt_jc5
# mediante su mtodo de separacin
indexList=['fragility_Qt_jc5'] #lista de la columna fragility_Qt_jc5
aggregator={'fragility_Qt': ['mean']} # este diccionario especifica la operacin a realizar para cada columna
#tabla  de la media de 'fragility_Qt' calculada y organizada por cada grupo en 'fragility_Qt_jc5'
pd.concat([theMapAndData[['fragility_Qt',col]].groupby(col,as_index=False).agg(aggregator) for col in indexList],axis=1)
```

```python
#creo un diccionario para fragility_Qt_jc5
newLabelsForLevels={0:"0_Great", 1:"1_Good", 2:"2_Middle", 3:"3_Bad", 4:"4_Poor"}
#creo la nueva columna y remplazo los valores de fragility_Qt_jc5 por los de newLabelsForLevels
theMapAndData['fragility_Qt_jc5_cat']=theMapAndData.loc[:,'fragility_Qt_jc5'].replace(newLabelsForLevels)

# visualizo las 20  primeras filas de la tabla
theMapAndData[['fragility_Qt','fragility_Qt_jc5','fragility_Qt_jc5_cat']].head(20)
```
