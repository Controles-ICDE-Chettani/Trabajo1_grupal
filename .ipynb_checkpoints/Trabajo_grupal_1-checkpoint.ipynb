```python
cities
```

```python
# añadimos información para cada país
import pandas as pd

fragilityCiaLink="https://github.com/CienciaDeDatosEspacial/merging/raw/main/FragilityCia_isos.csv"

fragilityCia=pd.read_csv(fragilityCiaLink)

fragilityCia.head()
```

```python
fragilityCia.replace(to_replace={'Country':try1},inplace=True)

# actualizamos los nombres de los países en el gragilityCia q no coinciden con countries
onlyFragilCia=set(fragilityCia.Country)- set(countries.COUNTRY)
onlyMap=set(countries.COUNTRY)- set(fragilityCia.Country)
# nuevas coincidencias
[(country, process.extractOne(country,onlyMap)) for country in sorted(onlyFragilCia)]
```

```python
# ahora algunos cambios manuales:
# aqui busca en countries los países con los siguientes códigos
countries[countries.COUNTRY.str.contains('LAO|SWA|KOR')]
```

```python
manualChanges={'SWAZILAND':'ESWATINI','LAOS':"LAO PEOPLE'S DEMOCRATIC REPUBLIC (THE)",'SOUTH KOREA':'KOREA (THE REPUBLIC OF)'}

countries.replace(to_replace={'COUNTRY':manualChanges},inplace=True)

# updating
onlyFragilCia=set(fragilityCia.Country)- set(countries.COUNTRY)
onlyMap=set(countries.COUNTRY)- set(fragilityCia.Country)
# new matches
[(country, process.extractOne(country,onlyMap)) for country in sorted(onlyFragilCia)]
```

```python
# estrategias de estandarización que nos permitirán que las variables tengan una escala comparable,
# lo cual facilita la comparación de sus distribuciones en el gráfico
```

```python
from sklearn.preprocessing import StandardScaler

#herramienta que transformará los datos para que tengan una media de 0 y una desviación estándar de 1.
scaler = StandardScaler()

#calcula la media y desviación estándar en DataNames y transforma los datos
normalized_data = scaler.fit_transform(theMapAndData[DataNames])
# fit_transform calcula la media y desviación estándar de las variables y los utiliza para transformar los datos

#nueva dataframe con datos normalizados, se pasa de ancho a largo y lo grafico en escala no logaritmica
sns.displot(pd.melt(pd.DataFrame(normalized_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
# técnica de escalado Min-Max, a diferencia del anterior, este también preserva la forma original de
# la distribución de los datos, pero en un rango específico de 0 a 1, en este caso de -0.2 a 1.2
from sklearn.preprocessing import MinMaxScaler

# valor mínimo se convierta en 0 y el valor máximo se convierta en 1.
scaler = MinMaxScaler(feature_range=(0, 1))

#calcular el rango mínimo y máximo de las columnas en DataNames y transformar
scaled_data=scaler.fit_transform(theMapAndData[DataNames])

#grafico de distribucion de un data frame nuevo con los datos transformados del DataNames pasado de ancho a largo
sns.displot(pd.melt(pd.DataFrame(scaled_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
# técnica de escalado robusto
from sklearn.preprocessing import RobustScaler
#En si todo el proceso normaliza las variables, haciendo menos susceptibles a influencias de valores atípicos.

#transformará los datos de manera que los valores estén centrados en la mediana y se escalen según el rango intercuartílico
scaler = RobustScaler()

#calcular la mediana y el rango intercuartílico de las columnas en DataNames y transforma datos
robScaled_data = scaler.fit_transform(theMapAndData[DataNames])

#grafico de distribucion de un data frame nuevo con los datos transformados del DataNames pasado de ancho a largo
sns.displot(pd.melt(pd.DataFrame(robScaled_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
#transformación de cuantiles:
from sklearn.preprocessing import QuantileTransformer
#En si transforma cuantiles a las variables, ajustándolas a una distribución normal.

#Los datos serán transformados en 99 rangos diferentes, se transformarán a una distribución normal/uniforme
scaler = QuantileTransformer(n_quantiles=99, random_state=0,output_distribution='normal')

#datos transformados en la distribución normal especificada.
QtScaled_data = scaler.fit_transform(theMapAndData[DataNames])

#Grafica distribucion de un data frame nuevo con los datos transformados del DataNames pasado de ancho a largo
sns.displot(pd.melt(pd.DataFrame(QtScaled_data,columns=DataNames)),
            x="value", hue="variable",kind="kde",
            log_scale=(False,False))
```

```python
#Nos quedamos con QtScaled_data (el ultimo)
theMapAndData['fragility_Qt']=QtScaled_data[:,0]
```

```python

import mapclassify #mapclassify para generar diferentes esquemas de clasificación según intervalos
import numpy as np

np.random.seed(12345) # so we all get the same results!
#para 5 intervalos, theVar asigno la columna fragility_Qt de theMapAndData
K=5
theVar=theMapAndData.fragility_Qt #esta es la fragilityqt
#EqualInterval---->Divide los valores de la variable en intervalos de igual tamaño
ei5 = mapclassify.EqualInterval(theVar, k=K)
#StdMean---->Crea intervalos basados en la media y la desviación estándar
msd = mapclassify.StdMean(theVar)
#Quantiles--->Crea intervalos donde cada uno contiene aproximadamente la misma cantidad de observaciones
q5=mapclassify.Quantiles(theVar,k=K)

#Basado en similitud, bueno para data multimodales
#detecta los puntos de ruptura entre los grupos de datos
mb5 = mapclassify.MaximumBreaks(theVar, k=K)
# Basado en similitud, bueno para datos asimetricos/asesgados
ht = mapclassify.HeadTailBreaks(theVar) # no K needed
# Basado en similitud, optimiza la homogeneidad dentro de los intervalos
# maximiza las diferencias entre rangos
fj5 = mapclassify.FisherJenks(theVar, k=K)
# Basado en similitud, maximiza las diferencias entre rangos
jc5 = mapclassify.JenksCaspall(theVar, k=K)
#Basado en similitud, Maximiza la homogeneidad dentro de los intervalos para datos sesgado
mp5 = mapclassify.MaxP(theVar, k=K)
```

```python
#lista de los diferentes metodos de clasificación de intervalos de la variable TheVar
class5 = ei5,msd, q5,mb5,  ht, fj5, jc5, mp5
#matriz de valores de ADCM de cada clasificador en class5.
#ADCM mide la desviación absoluta desde la mediana de cada intervalo
fits = np.array([ c.adcm for c in class5])
# pasamos ADCM a dataFrame
adcms = pd.DataFrame(fits)
# add columna classifier de adcms
adcms['classifier'] = [c.name for c in class5]
# Add column names to the ADCM
adcms.columns = ['ADCM', 'Classifier']
```

```python
#muestro el grafico generado
adcms.sort_values('ADCM').plot.barh(x='Classifier')
# ADCM significa calculo de la desviacion estandar, el grafico es sd de cada tipo de agrupacion de intervalo q se ha dado
```

```python
# guardamos el mejor método de clasificación en una nueva columna llamada fragility_Qt_jc5
theMapAndData['fragility_Qt_jc5'] = jc5.yb
```

```python
# datos originales con fragility_Qt y datos clasificados con fragility_Qt_jc5
theMapAndData[['fragility_Qt','fragility_Qt_jc5']].head()
```

```python
# este código tiene como objetivo general la media de fragility_Qt para cada categoria fragility_Qt_jc5
# mediante su método de separación
indexList=['fragility_Qt_jc5'] #lista de la columna fragility_Qt_jc5
aggregator={'fragility_Qt': ['mean']} # este diccionario especifica la operación a realizar para cada columna
#tabla  de la media de 'fragility_Qt' calculada y organizada por cada grupo en 'fragility_Qt_jc5'
pd.concat([theMapAndData[['fragility_Qt',col]].groupby(col,as_index=False).agg(aggregator) for col in indexList],axis=1)
```

```python
#creo un diccionario para fragility_Qt_jc5
newLabelsForLevels={0:"0_Great", 1:"1_Good", 2:"2_Middle", 3:"3_Bad", 4:"4_Poor"}
#creo la nueva columna y remplazo los valores de fragility_Qt_jc5 por los de newLabelsForLevels
theMapAndData['fragility_Qt_jc5_cat']=theMapAndData.loc[:,'fragility_Qt_jc5'].replace(newLabelsForLevels)

# visualizo las 20  primeras filas de la tabla
theMapAndData[['fragility_Qt','fragility_Qt_jc5','fragility_Qt_jc5_cat']].head(20)
```

```python
# crea un mapa coroplético que representa la fragilidad de diferentes regiones utilizando la
# columan fragility_Qt_jc5_cat del dataframe theMapAndData
import matplotlib.pyplot as plt

# f es la figura del mapa, el ax es el eje, figsize representa el tamaño (10x10pulgadas)
f, ax = plt.subplots(1, figsize=(10, 10))
theMapAndData.plot(column='fragility_Qt_jc5_cat', # variable to plot
                   cmap='viridis', # set of colors
                   categorical=True, # can be interpreted as category
                   edgecolor='white', # border color
                   linewidth=0., # width of border
                   alpha=1, # level of transparency (0 is invisible)
                   legend=True, # need a legend?
                   # location of legend: 'best', 'upper right', 'upper left', 'lower left',
                   # 'lower right', 'right', 'center left', 'center right',
                   # 'lower center', 'upper center', 'center'
                   legend_kwds={'loc':"lower left"},
        ax=ax
       )

ax.set_axis_off()
```
